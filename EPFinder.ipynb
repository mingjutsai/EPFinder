```python
from supervised.automl import AutoML
import joblib
import pandas as pd
from sklearn.metrics import (
    roc_auc_score, 
    average_precision_score,
)
```

# GM12878

```python
gm12878 = pd.read_csv("dataset/gm12878.scale.input.tsv", sep='\t')
X_gm12878 = gm12878.drop(['Class'], axis=1)
y_gm12878 = gm12878['Class']
```

```python
EPFinder_model = joblib.load('EPFinder_model.pkl')
```

```python
y_gm12878_pred_prob = EPFinder_model.predict_proba(X_gm12878)
gm12878_prob = pd.DataFrame(y_gm12878_pred_prob)
gm12878['EPFinder_Prediction'] = gm12878_prob[1]

print("\n" + "="*80)
print("THRESHOLD-INDEPENDENT METRICS")
print("="*80)

y_true = gm12878['Class'].values
y_score = gm12878['EPFinder_Prediction'].values

# AUROC (Area Under ROC Curve)
auroc = roc_auc_score(y_true, y_score)

# AUPRC (Area Under Precision-Recall Curve)
auprc = average_precision_score(y_true, y_score)

print(f"\n{'='*60}")
print(f"AUROC (Area Under ROC Curve):           {auroc:.4f}")
print(f"AUPRC (Area Under Precision-Recall):    {auprc:.4f}")
print(f"{'='*60}")
```
