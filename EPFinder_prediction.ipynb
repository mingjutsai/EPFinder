```python
import pandas as pd
from sklearn.metrics import (
    roc_auc_score, 
    average_precision_score,
)
from pycaret.classification import *
```

```python
EPFinder_model = load_model('finalize_EPFinder_model')
```

```python
EPFinder_model
```

# GM12878 validataion

```python
gm12878 = pd.read_csv("./dataset/gm12878_29features_ML.csv")
```

```python
gm12878.to_csv('./dataset/gm12878_29features_ML.tsv',sep='\')
```

```python
gm12878_29features = gm12878.drop(['#Class','ID','Enh','Prom','TX'],axis=1)
```

```python
gm12878_29features
```

```python
gm12878_predict = predict_model(EPFinder_model, raw_score = True, data = gm12878_29features)
```

```python
gm12878_predict
```

```python
gm12878['EPFinder_Prediction'] = gm12878_predict['prediction_score_1']

print("\n" + "="*80)
print("THRESHOLD-INDEPENDENT METRICS")
print("="*80)

y_true = gm12878['#Class'].values
y_score = gm12878['EPFinder_Prediction'].values

# AUROC (Area Under ROC Curve)
auroc = roc_auc_score(y_true, y_score)

# AUPRC (Area Under Precision-Recall Curve)
auprc = average_precision_score(y_true, y_score)

print(f"\n{'='*60}")
print(f"AUROC (Area Under ROC Curve):           {auroc:.4f}")
print(f"AUPRC (Area Under Precision-Recall):    {auprc:.4f}")
print(f"{'='*60}")
```
